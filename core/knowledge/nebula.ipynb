```python
import pandas as pd
from core.tools.nebula import Nebula
from config import get_cfg
```

```python
cfg = get_cfg()
nebula = Nebula(cfg)
```

```python
cat = pd.read_csv('data/knowledge_graph/chat_cn.csv')
cat.columns = ["_".join(x.strip().lower().split(' ')) if ' ' in x else x.strip().lower() for x in cat.columns]
cat = cat.rename(columns={'cold-resistant': 'cold_resistant', 
                          'hot-resistant':'hot_resistant',
                          'body.1': 'body_size',
                          'disqualify': 'disqualifing_faults', 
                          'foreface': 'facial_region', 
                          'penalize': 'faults', 
                          'colors': 'colour',
                          'pros&cons': 'pros_cons',
                          'feet': 'foot'}).fillna('')

dog = pd.read_csv('data/knowledge_graph/chien_cn.csv')
dog = dog.rename(columns={'cold-resistant': 'cold_resistant', 
                          'hot-resistant':'hot_resistant',
                          'jaw': 'jaws',
                          'forearms': 'forearms',
                          'n.b.': 'note',
                          'colour_type': 'colour',
                          'total_intro': 'root_intro',
                          'pros&cons': 'pros_cons',
                          'total_img': 'root_img',
                          'body': 'torso',
                          'hair_intro': "coat_intro", 'hair_img':'coat_img',
                          'front_intro': 'forequarters_intro', 'front_img': 'forequarters_img', 
                          'back_intro': 'hindquarters_intro', 'back_img' : 'hindquarters_img',
                          'eye_intro':'eyes_intro', 'eye_img':'eyes_img', 
                          'ear_intro':'ears_intro', 'ear_img':'ears_img'}).fillna('')

parts = pd.read_csv('data/knowledge_graph/part.csv').fillna('')
parts.columns = ["_".join(x.strip().lower().split(' ')) if ' ' in x else x.strip().lower() for x in parts.columns]
parts['is_parts_of'] = parts['is_parts_of'].apply(lambda x: x.lower().strip())
parts['english_name'] = parts['english_name'].apply(lambda x: x.lower().strip())
parts2attr = parts[parts['is_attr_of'] != ''].groupby('is_attr_of')['english_name'].apply(lambda x: ",".join(x)).to_dict()

```

```python
habit_dict = {'stick_people': '粘人程度', 'calm': '冷静程度', 'protect': '守卫能力', 'independent': '独立程度', 
              'chase': '狩猎能力', 'noisy': '吵闹程度', 'child_friendly': '小孩友善程度', 'animal_friendly': '动物友善程度', 
              'stranger_friendly': '生人友善程度', 'trainable': '训练程度', 'obedience': '听话程度', 
              'suitable_appartment': '适合公寓', 'suitable_new_master': '适合新主人', 'hot_resistant': '耐热程度', 
              'cold_resistant': '耐寒程度', 'disease_resistant': '抵御疾病能力', 'easy_gain_weight': '易增重', 
              'easy_maintance': '易保养程度', 'maintanance_cost': '保养成本', 'hair_loss': '掉毛程度', 
              'slobber': '口水程度', 'cleanliness': '干净程度', 'sport_need': '运动需求', 'energy': '精力水平', 
              'playfulness': '爱玩程度', 'beauty': '美容需要', 'smell': '体味程度'}

dog_attr = ['chinese_name', 'english_name', 'foreign_alias', 'chinese_alias', 'rate', 'hair_length', 
            'origin', 'body_size', 'head_form', 'male_height', 'male_weight', 'female_height', 
            'female_weight', 'life', 'price', 'cost_yearly', 'origin_function', 'function', 'IQ', 
            'intro_chinese', 'nature', 'pros_cons', 'feed_method', 'identification', 'image', 'suitable_owner_type']

cat_attr = ['chinese_name', 'english_name', 'foreign_alias', 'chinese_alias', 'hair_length', 'rate', 
            'origin', 'male_height', 'male_weight', 'female_height', 'female_weight', 'life', 'price', 
            'cost_yearly', 'body_size', 'intro_chinese', 'nature', 'pros_cons', 'feed_method', 'identification', 
            'image']
dog_intro = ['root_intro', 'root_img', 'coat_intro', 'coat_img', 'head_intro', 'head_img', 
              'eyes_intro', 'eyes_img', 'ears_intro', 'ears_img', 'nose_intro', 'nose_img', 'tail_intro', 'tail_img', 
              'chest_intro', 'chest_img', 'neck_intro', 'neck_img', 'forequarters_intro', 'forequarters_img', 
             'hindquarters_intro', 'hindquarters_img']

faults = ['faults', 'severe_faults', 'disqualifing_faults']

tica_info = ['group', 'category', 'divisions', 'permissible_outcrosses']

fci_info = ['name', 'english_name', 'france_name', 'german_name', 'spanish_name', 'FCI_NO', 'group', 'section', 'note']
```

```python
nebula.query('USE PetMedical;')

nebula.query('INSERT VERTEX association(name) VALUES "association001":("{}");'.format('FCI'))
nebula.query('INSERT VERTEX association(name) VALUES "association002":("{}");'.format('TICA'))
```

```python
cnt = 100
habit2hid = {}
for k, v in habit_dict.items():
    hid = "habit{}".format(str(cnt))
    habit2hid[k] = hid
    nebula.query('INSERT VERTEX habit(name, explaination) VALUES "{}":("{}", "{}");'.format(hid ,k, v))
#     print(nebula.query('INSERT VERTEX habit(name, explaination) VALUES "habit{}":("{}", "{}");'.format(str(cnt),k, v)))
    cnt += 1
```

```python
cnt = 100
parts2pid = {}
for idx, row in parts[parts['is_attr_of'] == ''].iterrows():
    pid = 'parts{}'.format(str(cnt))
    
    part = row['english_name'].lower().strip()
    parts2pid[part] = pid
    nebula.query('INSERT VERTEX body_parts(name, english_name) VALUES "{}":("{}", "{}");'.format(pid, row['name'], part))
    cnt += 1
    if row['is_parts_of'] != '':
        for item in row['is_parts_of'].split(','):
            nebula.query('INSERT EDGE is_part_of() VALUES "{}" -> "{}":();'.format(pid, parts2pid[item.strip()]))
```

```python
cnt = 1
cat2pid = {}
for idx, row in cat.iterrows():
    pid = 'cat{}'.format("0" * (4 - len(str(cnt))) + str(cnt))
    cat2pid[row['chinese_name']] = pid
    name = row['chinese_name']
    (nebula.query('INSERT VERTEX cat(chinese_name) VALUES "{}":("{}");'.format(pid, name)))
    for col in [x for x in cat_attr if x != 'chinese_name']:
        if 'height' in col:
            value = "-".join(row[col].split('_')) + 'cm'
        elif 'weight' in col:
            value = "-".join(row[col].split('_')) + 'kg'
        elif 'weight' in col:
            value = "-".join(row[col].split('_')) + 'kg'
        elif col in ['life', 'price', 'cost_yearly']:
            value = "-".join(row[col].split('_'))
        elif col == 'rate':
            try:
                value = float(row[col])
            except:
                value = ''
        else:
            value = row[col].replace('\n', '|')
        if value:
            if isinstance(value, float):
                nebula.query('UPSERT VERTEX ON cat "{}" SET {} = {} WHEN chinese_name == "{}";'.format(pid, col, value, name))
            else:
                nebula.query('UPSERT VERTEX ON cat "{}" SET {} = "{}" WHEN chinese_name == "{}";'.format(pid, col, value, name))
    
    cnt += 1
    for col in faults:
        if col in row.index:
            if col == 'faults':
                content = row['faults'].replace('\n', '|')
            elif col == 'severe_faults':
                content = row['severe_faults'].replace('\n', '|')
            elif col == 'disqualifing_faults':
                content = row['disqualifing_faults'].replace('\n', '|')
            else:
                content = ''
            if content:
                nebula.query('INSERT EDGE has_disqualifing_faults(content) VALUES "{}" -> "{}":("{}");'.format(pid, 'association002', content))
    
    for col in habit_dict.keys():
        if col in row.index and row[col] != '':
            nebula.query('INSERT EDGE habit_score(score) VALUES "{}" -> "{}":({});'.format(pid, habit2hid[col], int(row[col])))
    
    nebula.query('INSERT EDGE tica_info(group, category, divisions, permissible_outcrosses) VALUES "{}" -> "{}":("{}", "{}", "{}", "{}");'.format(pid, 'association002', *row[tica_info].values.tolist()))
    
    for col in parts[parts['is_attr_of'] == '']['english_name'].values:
        if col in row.index:
            if col == 'head':
                content = "|".join([x for x in row[[x for x in parts2attr[col].split(',') if x in row.index]].values.tolist() if x]).strip()
            elif col == 'coat':
                content = "|".join([x for x in row[[x for x in parts2attr[col].split(',') if x in row.index]].values.tolist() if x]).strip()
            else:
                content = row[col].strip().replace('\n', '|')
            if content:
                nebula.query('INSERT EDGE tica_standard(content) VALUES "{}" -> "{}":("{}");'.format(pid, parts2pid[col], content))
    
```

```python
cnt = 1
dog2pid = {}
for idx, row in dog.iterrows():
    pid = 'dog{}'.format("0" * (4 - len(str(cnt))) + str(cnt))
    dog2pid[row['chinese_name']] = pid

    name = row['chinese_name']
    (nebula.query('INSERT VERTEX dog(chinese_name) VALUES "{}":("{}");'.format(pid, name)))
    for col in [x for x in dog_attr if x != 'chinese_name']:
        if 'height' in col:
            value = "-".join(row[col].split('_')) + 'cm'
        elif 'weight' in col:
            value = "-".join(row[col].split('_')) + 'kg'
        elif 'weight' in col:
            value = "-".join(row[col].split('_')) + 'kg'
        elif col in ['life', 'price', 'cost_yearly']:
            value = "-".join(row[col].split('_'))
        elif col == 'rate':
            try:
                value = float(row[col])
            except:
                value = ''
        else:
            value = row[col].replace('\n', '|')
        if value:
            if isinstance(value, float):
                nebula.query('UPSERT VERTEX ON dog "{}" SET {} = {} WHEN chinese_name == "{}";'.format(pid, col, value, name))
            else:
                nebula.query('UPSERT VERTEX ON dog "{}" SET {} = "{}" WHEN chinese_name == "{}";'.format(pid, col, value, name))
       
    cnt += 1
    for col in faults:
        if col in row.index:
            if col == 'faults':
                content = row['faults'].replace('\n', '|')
            elif col == 'severe_faults':
                content = row['severe_faults'].replace('\n', '|')
            elif col == 'disqualifing_faults':
                content = row['disqualifing_faults'].replace('\n', '|')
            else:
                content = ''
            if content:
                nebula.query('INSERT EDGE has_disqualifing_faults(content) VALUES "{}" -> "{}":("{}");'.format(pid, 'association002', content))
   
    for col in habit_dict.keys():
        if col in row.index and row[col] != '':
            nebula.query('INSERT EDGE habit_score(score) VALUES "{}" -> "{}":({});'.format(pid, habit2hid[col], int(row[col])))
    
    nebula.query('INSERT EDGE fci_info(name, english_name, france_name, german_name, spanish_name, FCI_NO, \
    group, section, note) VALUES "{}" -> "{}":("{}", "{}", "{}", "{}", "{}", "{}", "{}", "{}"\
    );'.format(pid, 'association001', *row[fci_info].values.tolist()))
    
    for col in parts[parts['is_attr_of'] == '']['english_name'].values:
        if col in row.index:
            if col == 'head':
                content = "|".join([x for x in row[[x for x in parts2attr[col].split(',') if x in row.index]].values.tolist() if x]).strip()
            elif col == 'coat':
                content = "|".join([x for x in row[[x for x in parts2attr[col].split(',') if x in row.index]].values.tolist() if x]).strip()
            else:
                content = row[col].strip().replace('\n', '|')
            if content:
                nebula.query('INSERT EDGE fci_standard(content) VALUES "{}" -> "{}":("{}");'.format(pid, parts2pid[col], content))
        
        if col + '_intro' in row.index and row[col + '_intro'] != '':
            (nebula.query('INSERT EDGE body_parts_info(intro, image) VALUES "{}" -> "{}":("{}", "{}");'.format(pid, parts2pid[col], row[col + '_intro'].replace('\n', '|'), row[col + '_img'])))
```
